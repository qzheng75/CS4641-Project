```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

def kfold_30_3_secs(df_30, df_3, nfold, shuffle_each_fold=True):
    idx_perm = np.random.permutation(np.arange(df_30.shape[0]))
    test_30_indices = [set(indices) for indices in np.array_split(idx_perm, nfold)]
    
    for i in test_30_indices:
        print(i)

    train_data, test_data = [], []
    for k in range(nfold):
        train_3sec_df = df_3[~df_3['30_sec_index'].isin(test_30_indices[k])]
        X_train, y_train = train_3sec_df.drop(['30_sec_index', 'label'], axis=1), train_3sec_df['label']
        
        if shuffle_each_fold:
            range_n = np.arange(len(y_train))
            idx_perm = np.random.permutation(range_n)
            X_train, y_train = X_train.iloc[idx_perm], y_train.iloc[idx_perm]
            
        k_test_indices = np.array(list(test_30_indices[k]))
        X_test, y_test = df_30.iloc[k_test_indices].drop(['audio_idx', 'label'], axis=1),\
            df_30.iloc[k_test_indices]['label']
        
        y_train = encoder.fit_transform(y_train)
        y_test = encoder.fit_transform(y_test)
        
        train_data.append((X_train, y_train))
        test_data.append((X_test, y_test))
    return train_data, test_data
```

```python
train_data, test_data = kfold_30_3_secs(df_30sec, df_3sec, 5)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit the scaler on the training set and transform all sets
X_train = scaler.fit_transform(train_data[0][0])
X_test = scaler.transform(test_data[0][0])
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
y_train = encoder.fit_transform(train_data[0][1])
y_test = encoder.fit_transform(test_data[0][1])
```

```python
import matplotlib.pyplot as plt
from audio_toolbox.metrics import calculate_acc, precision_recall, visualize_confusion_matrices

def fit_and_evaluate_model(model, X_train, y_train, X_test, y_test, model_name='Model'):
    model.fit(X_train, y_train)
    print('Model fitting finished')

    conf_mats = []
    for x, y, split in zip([X_train, X_test],
                    [y_train, y_test],
                    ['Train', 'Test']):
        acc, correct, incorrect = calculate_acc(model, x, y)
        print(f'{split} accuracy: {acc:.4f}%, {len(incorrect)} mismatches out of {len(incorrect) + len(correct)} samples')
        conf_mat, precision, recall, f1 = precision_recall(model, x, y, return_each_class=False)
        conf_mats.append(conf_mat)
        print(f'(Averaged) {split} precision: {precision:.4f}, recall: {recall:.4f}, f1 score: {f1:.4f}')
        
    visualize_confusion_matrices(conf_mats,
                                titles=('Train', 'Test'),
                                grand_title=f'Confusion matrices for {model_name}')
```

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression(max_iter=1000, random_state=RANDOM_STATE)
fit_and_evaluate_model(logistic_model, X_train, y_train, X_test, y_test, 'Logistic Regression')
```

```python
from sklearn.svm import SVC

svm_classifier = SVC(kernel='rbf', random_state=RANDOM_STATE)
fit_and_evaluate_model(svm_classifier, X_train, y_train, X_test, y_test, 'Linear SVM')
```

```python
from sklearn.svm import SVC

ovo_svm = SVC(kernel='rbf', decision_function_shape='ovo', C=2, random_state=RANDOM_STATE)
fit_and_evaluate_model(ovo_svm, X_train, y_train, X_test, y_test, 'ovo SVM')
```

```python
from sklearn.ensemble import RandomForestClassifier

rand_forest_classifier = RandomForestClassifier(max_depth=15,
                                                max_features=25,
                                                random_state=RANDOM_STATE)
fit_and_evaluate_model(rand_forest_classifier, X_train, y_train, X_test, y_test, 'Random Forest')
```

```python
from sklearn.ensemble import GradientBoostingClassifier

xgboost_classifier = GradientBoostingClassifier(max_depth=10, max_features=10, random_state=RANDOM_STATE)
fit_and_evaluate_model(xgboost_classifier, X_train, y_train, X_test, y_test, 'XGBoost')
```

```python
current_models = [logistic_model, svm_classifier, ovo_svm, rand_forest_classifier]
current_model_names = ["Logistic", "SVM", "OVO SVM", "Random Forest"]
```

```python
from audio_toolbox.metrics import kfold_validation
import warnings

N_SPLITS = 5

accs_dict = {}
f1s_dict = {}

accs = []
precisions = []
recalls = []
f1s = []

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    for model, name in zip(current_models, current_model_names):
        print(f"Benchmarking {name}")
        
        for train, test in zip(train_data, test_data):
            train = X_train, y_train
            test = X_test, y_test
            X_train = scaler.transform(X_train)
            model.fit(X_train, y_train)
            X_test = scaler.transform(X_test)
            
            acc, _, _ = calculate_acc(model, X_test, y_test)
            _, precision, recall, f1 = precision_recall(model, X_test, y_test, return_each_class=False)
            accs.append(acc)
            precisions.append(precision)
            recalls.append(recall)
            f1s.append(f1)
            
        avg_acc, avg_precision, avg_recall, avg_f1 = np.mean(accs), np.mean(precisions), np.mean(recalls), np.mean(f1s)
        accs_dict[name] = avg_acc
        f1s_dict[name] = avg_f1
        print(f"Average Accuracy: {avg_acc:.4f}")
        print(f"Average Precision: {avg_precision:.4f}")
        print(f"Average Recall: {avg_recall:.4f}")
        print(f"Average F1 Score: {avg_f1:.4f}")
        print()
```
